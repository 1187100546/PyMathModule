```python
from sympy import integrate, log, exp, symbols, derive_by_array, lambdify
import matplotlib.pyplot as plt
import numpy as np
```

```python
# lambdify?
# np.linspace?
# plt.text?
```

```python
x = symbols('x')
```

```python
sigmoid = 1/(1+exp(-x))
```

```python
sigmoid_derive = derive_by_array(sigmoid, x)
```

```python
# 给定x计算y
sigmoid_derive.evalf(subs={x:0})
```

```python
# 使用lambdify转换为函数
sigmoid_function = lambdify(x, sigmoid, 'numpy')
```

```python
sigmoid_derive_function = lambdify(x, sigmoid_derive, 'numpy')
```

```python
x_linspaces = np.linspace(-5, 5, 100)
```

```python
y_linspaces = sigmoid_derive_function(x_linspaces)
```

```python
plt.plot(x_linspaces, y_linspaces)
plt.title('sigmoid derive')
```

```python
y_linspaces = sigmoid_function(x_linspaces)
```

```python
plt.plot(x_linspaces, y_linspaces)
plt.title('sigmoid')
```
